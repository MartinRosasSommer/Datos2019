```python
import pandas as pd
```

```python
auctions = pd.read_csv("auctions.csv")
```

```python
installs = pd.read_csv("installs.csv")
```

```python
auctions = auctions[["date", "device_id"]]
```

```python
auctions["is_install"] = False
```

```python
installs.head()
```

```python
installs = installs[["created", "ref_hash"]]
```

```python
installs["is_install"] = True
```

```python
installs = installs.rename(index=str,columns={"created": "date", "ref_hash": "device_id"})
```

```python
installs.head()
```

```python
auctions.head()
```

```python
concatenate_installs_auctions = pd.concat([installs, auctions])
```

```python
ordered = concatenate_installs_auctions.sort_values(by=["device_id",'date'])
```
