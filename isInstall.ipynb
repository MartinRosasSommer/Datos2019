### Feature cantidad de installs y auction por dispositivo

```python
auctions["Cantidad_total_de_auction"] = 1
```

```python
cantidad_total_auctions = auctions.groupby("device_id").agg({"Cantidad_total_de_auction":"sum"}).reset_index()
cantidad_total_auctions
```

```python
auctions.loc[((auctions["date"] < "2019-04-21")), "Periodo"] = "Primero"
auctions.loc[((auctions["date"] > "2019-04-20") & (auctions["date"] < "2019-04-24") ), "Periodo"] = "Segundo"
auctions.loc[((auctions["date"] > "2019-04-23")), "Periodo"] = "Tercero"
```

```python
auctions_periodos = auctions.groupby(["device_id","Periodo"]).agg({"Cantidad_total_de_auction":"sum"})
```

```python
auctions_periodos = auctions_periodos.unstack().reset_index()
auctions_periodos.fillna(0, inplace= True)
```

```python
auctions_periodos.columns = ["device_id","1_Periodo_Auctions","2_Periodo_Auctions","3_Periodo_Auctions"]
```

```python
auctions_periodos
```

```python
#Joineo 
auctions_cantidades = auctions_periodos.merge(cantidad_total_auctions, how = "inner", on = "device_id")
```

### Installs

```python
installs = pd.read_csv("installs.csv")
```

```python
installs = installs.rename(index=str,columns={"created": "date", "ref_hash": "device_id"})
```

```python
installs["Cantidad_total_de_install"] = 1
```

```python
cantidad_total_installs = installs.groupby("device_id").agg({"Cantidad_total_de_install":"sum"}).reset_index()
cantidad_total_installs
```

```python
installs.loc[((installs["date"] < "2019-04-21")), "Periodo"] = "Primero"
installs.loc[((installs["date"] > "2019-04-20") & (installs["date"] < "2019-04-24") ), "Periodo"] = "Segundo"
installs.loc[((installs["date"] > "2019-04-23")), "date"] = "Tercero"
```

```python
installs_periodos = installs.groupby(["device_id","Periodo"]).agg({"Cantidad_total_de_install":"sum"})
```

```python
periodos_installs = installs_periodos.unstack().reset_index()
periodos_installs.fillna(0, inplace= True)
```

```python
periodos_installs.columns = ["device_id","1_Periodo_Installs","2_Periodo_Installs","3_Periodo_Installs"]
```

```python
#Joineo 
installs_cantidades = periodos_installs.merge(cantidad_total_installs, how = "inner", on = "device_id")
```

```python
installs_cantidades
```

### Mergeo installs con auctions

```python
auctions_installs_cantidades = installs_cantidades.merge(auctions_cantidades, how = "inner", on = "device_id")
```
