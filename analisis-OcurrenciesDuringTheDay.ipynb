```python
import pandas as pd
import matplotlib.pyplot as plt
parse_dates_clicks = ["created"]
clicks = pd.read_csv('clicks.csv',parse_dates=parse_dates_clicks)
parse_dates_auctions = ["date"]
auctions = pd.read_csv('auctions.csv', parse_dates=parse_dates_auctions)
parse_dates_installs = ["created"]
installs = pd.read_csv('installs.csv', parse_dates=parse_dates_installs)
parse_dates_events = ["date"]
events = pd.read_csv('events.csv', parse_dates=parse_dates_events)
```

```python
clicks["created"] = pd.to_datetime(clicks["created"], utc=True)
```

```python
clicks["created"] = clicks["created"].dt.tz_convert("America/Buenos_Aires")
```

```python
clicksIn30MinutesPeriods = clicks.groupby(["created"])["created"].count()
```

```python
auctions["date"] = pd.to_datetime(auctions["date"], utc=True)
```

```python
installsIn30MinutesPeriods = installs.groupby(["created"])["created"].count()
```

```python
eventsRounded = events.copy()
```

```python
eventsRounded["date"] = pd.to_datetime(eventsRounded["date"], utc=True)
```

```python
eventsRounded["date"] = eventsRounded["date"].dt.tz_convert("America/Buenos_Aires")
```

```python
eventsRounded["date"] = pd.to_datetime(eventsRounded["date"]).dt.floor('30T').dt.time
```

```python
eventsIn30MinutesPeriods = eventsRounded.groupby(["date"])["date"].count()
```

```python
eventsIn30MinutesPeriods.plot(title="Events during the day")
```

```python
auctionsIn30MDF = auctionsIn30MinutesPeriods.to_frame()
auctionsIn30MDF.columns = ["count"]
auctionsIn30MDF.reset_index()
auctionsIn30MDF['min'] = auctionsIn30MDF["count"][(auctionsIn30MDF["count"].shift(1) > auctionsIn30MDF["count"]) & (auctionsIn30MDF["count"].shift(-1) > auctionsIn30MDF["count"])]
auctionsIn30MDF['max'] = auctionsIn30MDF["count"][(auctionsIn30MDF["count"].shift(1) < auctionsIn30MDF["count"]) & (auctionsIn30MDF["count"].shift(-1) < auctionsIn30MDF["count"])]
```

```python
plt.scatter(auctionsIn30MDF.index, auctionsIn30MDF['min'], c='r')
plt.scatter(auctionsIn30MDF.index, auctionsIn30MDF['max'], c='g')
auctionsIn30MDF["count"].plot()
```

```python
auctionsIn30MDF[auctionsIn30MDF["max"].notnull() | auctionsIn30MDF["min"].notnull()]
```

```python
installsIn30MDF = installsIn30MinutesPeriods.to_frame()
installsIn30MDF.columns = ["count"]
installsIn30MDF.reset_index()
installsIn30MDF['min'] = installsIn30MDF["count"][(installsIn30MDF["count"].shift(1) > installsIn30MDF["count"]) & (installsIn30MDF["count"].shift(-1) > installsIn30MDF["count"])]
installsIn30MDF['max'] = installsIn30MDF["count"][(installsIn30MDF["count"].shift(1) < installsIn30MDF["count"]) & (installsIn30MDF["count"].shift(-1) < installsIn30MDF["count"])]
```

```python
plt.scatter(installsIn30MDF.index, installsIn30MDF['min'], c='r')
plt.scatter(installsIn30MDF.index, installsIn30MDF['max'], c='g')
installsIn30MDF["count"].plot()
```

```python
installsIn30MDF[installsIn30MDF["max"].notnull() | installsIn30MDF["min"].notnull()]
```
