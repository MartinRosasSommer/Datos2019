```python
eventsDatesByIpAddressFloored = pd.Series(eventsDatesByIpAddress["date"]).map(lambda x : pd.Series(x).map(lambda y : y.floor('30S')).tolist()).tolist()
```

```python
eventsDatesByIpAddressFloored
```

```python
eventsDatesByIpAddress["date"] = eventsDatesByIpAddressFloored
```

```python
eventsDatesByIpAddressFlooredDuplicateCounts = eventsDatesByIpAddressFloored
```

```python
eventsDatesByIpAddressFlooredDuplicateCounts = pd.Series(eventsDatesByIpAddressFlooredDuplicateCounts).to_frame()
```

```python
eventDatesOcurrences = []
for x in eventsDatesByIpAddressFlooredDuplicateCounts[0]:
    auxDict = {}
    eventDatesOcurrences.append(auxDict)
    for date in x:
        if date not in auxDict:
            auxDict[date] = 1
        else:
            auxDict[date] += 1
```

```python
eventDatesOcurrences
```

```python
eventsDatesByIpAddress["date"] = eventDatesOcurrences
```

```python
len(eventsDatesByIpAddress.index)
```

```python
#Filtro los que tengan mÃ¡s de 4 eventos en menos de 30 secs
filteredEvents = pd.DataFrame({'ip': [], 'date': []})
count = 0;
while count != 285212:
    currentRow = eventsDatesByIpAddress.iloc[count]
    add = True;
    for i in currentRow["date"].values():
        if (i >= 4):
            add = False;
            break;
    if add:
        filteredEvents = filteredEvents.append({"ip": currentRow["ip_address"], "date": currentRow["date"]}, ignore_index=True)
    count += 1
```

```python
#Vuelvo a sumar la cantidad de eventos por ip
filteredEventsIpsCounts = pd.DataFrame({'ip': [], 'count': []})
count = 0;
while count != 285212:
    currentRow = eventsDatesByIpAddress.iloc[count]
    auxCount = 0
    for i in currentRow["date"].values():
        auxCount += i;
    
    filteredEventsIpsCounts = filteredEventsIpsCounts.append({"ip": currentRow["ip_address"], "count": auxCount}, ignore_index=True)
    count += 1
```
