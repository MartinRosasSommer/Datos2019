```python
import pandas as pd
parse_dates_events = ["date"]
events = pd.read_csv('events.csv.gzip',  compression = 'gzip', parse_dates=parse_dates_events)
```

```python
events.head()
```

```python
eventsDatesByIpAddress = events.groupby(['ip_address']).agg({"date" : lambda x: x.tolist()})
```

```python
eventsDatesByIpAddress = eventsDatesByIpAddress.reset_index()
```

```python
eventsDatesByIpAddress[eventsDatesByIpAddress['ip_address'] == 1992632945768888579]["date"].tolist()
```

```python
eventsDatesByIpAddress["date"]
```

```python
eventsDatesByIpAddressFloored = pd.Series(eventsDatesByIpAddress["date"]).map(lambda x : pd.Series(x).map(lambda y : y.floor('30S')).tolist()).tolist()
```

```python
eventsDatesByIpAddressFloored
```

```python
eventsDatesByIpAddress["date"] = eventsDatesByIpAddressFloored
```

```python
eventsDatesByIpAddressFlooredDuplicateCounts = eventsDatesByIpAddressFloored
```

```python
eventsDatesByIpAddressFlooredDuplicateCounts = pd.Series(eventsDatesByIpAddressFlooredDuplicateCounts).to_frame()
```

```python
eventDatesOcurrences = []
for x in eventsDatesByIpAddressFlooredDuplicateCounts[0]:
    auxDict = {}
    eventDatesOcurrences.append(auxDict)
    for date in x:
        if date not in auxDict:
            auxDict[date] = 1
        else:
            auxDict[date] += 1
```

```python
eventDatesOcurrences
```

```python
eventsDatesByIpAddress["date"] = eventDatesOcurrences
```
